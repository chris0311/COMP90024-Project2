```python
import pandas as pd
import json

with open('sa2.json', 'r') as file:
    age = json.load(file)

sa2_data = []
for feature in age['features']:
    sa2_code = feature['properties']['sa2_code_2021']
    median_age = feature['properties']['median_age_persons']
    sa2_data.append({'SA2_CODE21': sa2_code, 'median_age': median_age})

sa2_df = pd.DataFrame(sa2_data)
sa2_df
```

```python
# turn sa2 df to a list of dict, including roww id
sa2_list = sa2_df.to_dict('records')
for i, row in enumerate(sa2_list):
    row['id'] = i

print(len(sa2_list))
sa2_list
    
```

```python
import datetime

def round_to_nearest_half_hour(dt):
    minute = dt.minute
    if minute < 15:
        return dt.replace(minute=0, second=0)
    elif minute < 45:
        return dt.replace(minute=30, second=0)
    else:
        return dt.replace(minute=0, second=0) + pd.Timedelta(hours=1)

# convert string to datetime
m_df['created_at'] = pd.to_datetime(m_df['created_at'])
m_df['rounded_created_at'] = m_df['created_at'].apply(round_to_nearest_half_hour)
m_df['rounded_created_at'] = m_df['rounded_created_at'].dt.tz_localize(None)
m_df
```

```python
bom_df['local_date_time'] = pd.to_datetime(bom_df['local_date_time'])
bom_df['rounded_local_date_time'] = bom_df['local_date_time'].apply(round_to_nearest_half_hour)
bom_df['rounded_local_date_time'] = bom_df['rounded_local_date_time'].dt.tz_localize(None)
bom_df
```

```python
merge_df = pd.merge(m_df, bom_df, left_on='rounded_created_at', right_on='rounded_local_date_time', how='inner')
merge_df
```

```python
from elasticsearch8 import Elasticsearch, helpers

client = Elasticsearch(
    'https://127.0.0.1:9200',
    verify_certs=False,
    ssl_show_warn=False,
    basic_auth=('elastic', 'elastic')
)

if client.ping():
    print("Connected to Elasticsearch")
else:
    print("Could not connect to Elasticsearch")
```

```python
index_name = 'sa2'
actions = [
    {
        "_index": index_name,
        "_id": doc["id"],
        "_source": doc
    }
    for doc in sa2_list
]

try:
    helpers.bulk(client, actions)
    print("Bulk upload successful")
except ConnectionError as e:
    print(f"Connection error: {e}")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
# Refresh the index to make the documents searchable
client.indices.refresh(index=index_name)

# Verify the documents were indexed
search_response = client.search(index=index_name)
print("Search response:", json.dumps(search_response.body, indent=2))
```

```python
from string import Template

days_expr= Template('''{
                    "range": {
                        "local_date_time": {
                            "gte": "${sdate}T00:00:00",
                            "lte": "${edate}T23:59:59"
                        }
                    }
                }''')

bom_expr = days_expr.substitute(sdate='2024-05-17', edate='2024-05-17')

res = client.search(
    index = 'observations',
    query = json.loads(bom_expr)
)

print(res['hits']['hits'])

```

```python
res1 = res['hits']['hits']
bom_lst = []
for i in res1:
    bom_lst.append(i['_source'])
bom_data = pd.DataFrame(bom_lst)
bom_data
```

```python
start_date  = "2024-05-17"
end_date = "2024-05-17"
size = 100

days_expr= Template('''{
                    "range": {
                        "created_at": {
                            "gte": "${sdate}T00:00:00",
                            "lte": "${edate}T23:59:59"
                        }
                    }
                }''')

query_template = Template('''{
    "query": ${days_expr},
    "size": ${size}
}''')

days_expr = days_expr.substitute(sdate=start_date, edate=end_date)
query = query_template.substitute(days_expr=days_expr, size=size)

res = client.search(
            index = 'mastodon',
            body = json.loads(query)
        )

print(res)
print(len(res['hits']['hits']))
```

```python
mres = res['hits']['hits']
m_lst = []
for i in mres:
    m_lst.append(i['_source'])
m_data = pd.DataFrame(m_lst)
m_data
```

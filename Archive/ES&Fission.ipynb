```python
import datetime

def round_to_nearest_half_hour(dt):
    minute = dt.minute
    if minute < 15:
        return dt.replace(minute=0, second=0)
    elif minute < 45:
        return dt.replace(minute=30, second=0)
    else:
        return dt.replace(minute=0, second=0) + pd.Timedelta(hours=1)

# convert string to datetime
m_df['created_at'] = pd.to_datetime(m_df['created_at'])
m_df['rounded_created_at'] = m_df['created_at'].apply(round_to_nearest_half_hour)
m_df['rounded_created_at'] = m_df['rounded_created_at'].dt.tz_localize(None)
m_df
```

```python
bom_df['local_date_time'] = pd.to_datetime(bom_df['local_date_time'])
bom_df['rounded_local_date_time'] = bom_df['local_date_time'].apply(round_to_nearest_half_hour)
bom_df['rounded_local_date_time'] = bom_df['rounded_local_date_time'].dt.tz_localize(None)
bom_df
```

```python
merge_df = pd.merge(m_df, bom_df, left_on='rounded_created_at', right_on='rounded_local_date_time', how='inner')
merge_df
```

```python
from sympy import Polygon, Point

def geojson_to_sympy_polygon(geojson):
    coordinates = geojson['coordinates'][0]
    points = [Point(float(coord[0]), float(coord[1])) for coord in coordinates[0]]
    return Polygon(*points)
```

```python
from sympy import Polygon, Point as SymPoint, Float, N
import random


response = client.search(
    index="geoindex",
    size=10, 
    body={
        "_source": ["location"],  # Fetch only the location field
        "query": {
            "match_all": {}
        }
    }
    )

hits = response['hits']['hits']

print("got response")
all_polygons = [geojson_to_sympy_polygon(doc['_source']['location']) for doc in hits]

# Combine all polygons (SymPy does not directly support MultiPolygon, so we'll assume they're all combined into one)
print("combining polygons")
combined_polygon = all_polygons[0]
for polygon in all_polygons[1:]:
    combined_polygon = combined_polygon.union(polygon)

print(len(all_polygons))
print(all_polygons)
```

```python
def generate_random_points_in_polygon(poly):
    minx, miny, maxx, maxy = poly.bounds
    while True:
        p = SymPoint(Float(random.uniform(float(minx), float(maxx))), Float(random.uniform(float(miny), float(maxy))))
        if poly.encloses_point(p):
            return p
```

```python
print(all_polygons)
    
# generate random points
rand_point = generate_random_points_in_polygon(combined_polygon)

# make point into ES format
location = {
    "lat": rand_point.y,
    "lon": rand_point.x
}

print(location)
```

```python
import numpy as np

def min_max_scale(data, feature_range=(-1, 1)):
    min_val = np.min(data, axis=0)
    max_val = np.max(data, axis=0)
    range_min, range_max = feature_range
    scaled_data = (data - min_val) / (max_val - min_val) * (range_max - range_min) + range_min
    return scaled_data

# Example usage
data = np.array([1,2,3])
scaled_data = min_max_scale(data, feature_range=(-1, 1))
print(scaled_data)
```

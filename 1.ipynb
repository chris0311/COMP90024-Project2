```python
file_path1 = './Rental.xlsx'
xls = pd.ExcelFile(file_path1)


all_properties_df = pd.read_excel(xls, sheet_name='All properties')

rental_prices = all_properties_df[['Unnamed: 1', 'Unnamed: 193']].copy()


rental_prices = rental_prices.dropna().iloc[2:]  
rental_prices.columns = ['Region', 'Median Price Dec 2023']  

rental_prices = rental_prices[rental_prices['Region'] != 'Group Total']

rental_prices.reset_index(drop=True, inplace=True)

rental_prices
```

```python
file_path2 = './Crime.xlsx'

data = pd.read_excel(file_path2, sheet_name='Table 03')

data_2023_dec = data[(data['Year'] == 2023) & (data['Year ending'] == 'December')]

crime_counts = data_2023_dec.groupby('Suburb/Town Name')['Offence Count'].sum().reset_index()

crime_counts
```

```python
merged_data = pd.merge(crime_counts, rental_prices, left_on='Suburb/Town Name', right_on='Region', how='inner')

merged_data
```

```python
merged_data['Median Price Dec 2023'] = pd.to_numeric(merged_data['Median Price Dec 2023'], errors='coerce')

correlation = merged_data['Offence Count'].corr(merged_data['Median Price Dec 2023'])

correlation
```
